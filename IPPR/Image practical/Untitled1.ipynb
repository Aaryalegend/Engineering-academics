```python
import cv2
import pytesseract
import numpy as np

# Initialize the video capture (use 0 for webcam or provide the file path)
cap = cv2.VideoCapture("car1.mp4")  # Or use "car1.mp4" for a video file

# Set up Tesseract for OCR (Ensure tesseract is installed and its path is set correctly)
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\IT\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'  # Windows example

# Function to process and recognize plates
def recognize_license_plate(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)  # Blur to reduce noise
    edges = cv2.Canny(blurred, 100, 200)  # Apply Canny edge detection

    # Find contours (potential license plates)
    contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contours:
        # Approximate the contour to find rectangular shapes
        epsilon = 0.04 * cv2.arcLength(cnt, True)
        approx = cv2.approxPolyDP(cnt, epsilon, True)

        # Check if the contour is rectangular (4 sides)
        if len(approx) == 4:
            # Get bounding box of the rectangle
            x, y, w, h = cv2.boundingRect(approx)
            aspect_ratio = w / float(h)

            # License plates are typically rectangles with certain aspect ratio (adjust as needed)
            if 2 < aspect_ratio < 5:  # Change values based on your requirements
                plate = frame[y:y+h, x:x+w]  # Extract license plate region
                plate_gray = cv2.cvtColor(plate, cv2.COLOR_BGR2GRAY)
                plate_text = pytesseract.image_to_string(plate_gray, config='--psm 8')  # OCR on plate

                # Display the plate region and its recognized text
                cv2.imshow("License Plate", plate)
                print("Detected License Plate: ", plate_text.strip())

                # Draw the contour and bounding box around the license plate
                cv2.drawContours(frame, [approx], -1, (0, 255, 0), 3)
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    return frame

# Main loop to process each frame from the video stream
while True:
    ret, frame = cap.read()
    
    if not ret:
        break

    # Call the license plate detection and recognition function
    frame_with_plate = recognize_license_plate(frame)

    # Display the video feed with license plate detection
    cv2.imshow("Real-time License Plate Recognition", frame_with_plate)

    # Exit on pressing the 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
