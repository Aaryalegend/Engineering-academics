```python
import cv2
import matplotlib.pyplot as plt

# Load the image in BGR (default color format in OpenCV)
img = cv2.imread('D://SJC//Image Processing//images//image1.jpg', cv2.IMREAD_COLOR)

# Convert the image to HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Convert the original image to RGB for Matplotlib (since OpenCV uses BGR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display images using Matplotlib
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img_rgb)  # Show the original image

plt.subplot(1, 2, 2)
plt.title("HSV Image")
plt.imshow(hsv_img)  # Convert back to RGB for display
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in RGB
img_rgb = cv2.imread('D://SJC//Image Processing//images//image1.jpg', cv2.IMREAD_COLOR)

# Convert the image from RGB to BGR (OpenCV uses BGR by default)
img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

# Convert the BGR image to LAB
lab_img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2LAB)

# Convert the original image to RGB for display
img_rgb_display = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Display the images
plt.subplot(1, 2, 1)
plt.title("Original RGB")
plt.imshow(img_rgb_display)

plt.subplot(1, 2, 2)
plt.title("Converted LAB")
# Since LAB image is in 3 channels, we can't display it directly in RGB format. 
# We will show only the L channel (brightness) for visualization purposes.
plt.imshow(lab_img[:, :, 0], cmap='gray')  # Show the L (luminance) channel as grayscale

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in RGB
img_rgb = cv2.imread('D://SJC//Image Processing//images//image1.jpg', cv2.IMREAD_COLOR)

# Convert the image from RGB to BGR (OpenCV uses BGR by default)
img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

# Convert the BGR image to YUV
yuv_img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2YUV)

# Convert the original image to RGB for display
img_rgb_display = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Display the images
plt.subplot(1, 2, 1)
plt.title("Original RGB")
plt.imshow(img_rgb_display)

plt.subplot(1, 2, 2)
plt.title("Converted YUV")
# Since YUV image has 3 channels (Y, U, V), we can visualize the Y channel as grayscale
plt.imshow(yuv_img[:, :, 0], cmap='gray')  # Show the Y (luminance) channel as grayscale

plt.show()
```
