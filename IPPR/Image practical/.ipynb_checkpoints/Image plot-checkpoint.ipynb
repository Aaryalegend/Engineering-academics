```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("D://SJC//Image Processing//tomato.jpg",1)
cv2.imshow('Tomato',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(0)  
plt.subplot(1, 2, 2) 
plt.title("original") 
plt.imshow(img)
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```
